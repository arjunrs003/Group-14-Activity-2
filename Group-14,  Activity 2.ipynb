```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Storing Data
df=pd.read_csv('Diabetes.csv')
```

```python
#Viewing few columns of data
df.head()
```

# 1. Create a count plot of outcome

```python
plt.title('Countplot of Outcome')
fig=sns.countplot(x='Outcome', data=df)

#to display the count 
for p in fig.patches:
    fig.annotate(p.get_height(), (p.get_x()+0.35, p.get_height()-150))
```

# To check chances of getting diabetes, if it is common in family using Diabetes pedigree function .

```python

sns.boxplot(x='Outcome', y='DiabetesPedigreeFunction', data=df)
```

Observation;- From the graph we can observe that,  chances of being diabetic is higher if it is common in family

# 2. To analyze the effect of Age on Outcome

```python
plt.title('Age v\s Outcome')
sns.boxplot(x='Outcome', y='Age', data=df)
```

observation:-Chances of getting diabetes increases as age increases

# To check for the relationship with insulin & Glucose and group by Outcome

```python
sns.scatterplot(x='Insulin',y='Glucose', data=df, hue='Outcome')
```

observation:- There is a linear relationship  between Insulin and Glucose

# To create a heatmap on correlation of features 

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```
